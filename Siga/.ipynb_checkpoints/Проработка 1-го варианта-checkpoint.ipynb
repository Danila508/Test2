```python
#Импорт библиотек
import numpy as np
import pandas as pd # импорт модуля Pandas
import matplotlib.pyplot as plt # импорт инструментов визуализации
import seaborn as sns # импорт пакета Seaborn для визуализации
```

```python
# получение данных из репозитория
df=pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

```python
# просмотр первых 5-ти записей
df.head()
```

```python
# просмотр общей информации о DataFrame: количество записей/строк, описание всех колонок, их названия, тип данных и описание не пустых значенией
df.info()
```

```python
# Приведение типов
df['date'] = df.date.astype("datetime64")
```

```python
df.info()
```

```python
# просмотр количества пропусков в каждом столбце
pd.set_option('display.max_rows',None) # Максимальную длину вывода данных увеличиваем до неограниенно, чтобы посмотреть количесвто пропусков по всем колонкам
df.isna().sum()
```

```python
# Просмотр количества строк и колонок
print('Размерность данных: количество строк - {0}, количество колонок - {1}'.format(df.shape[0], df.shape[1]))
```

```python
# Заполнение пустых полей
df[['new_cases', 'new_deaths']]=df[[ 'new_cases', 'new_deaths']].fillna(0)
```

```python
# Проверка заменились пустые значения или нет
df.isna().sum()
```

```python
# Формирование дополнительных атрибутов
df['year'] = df['date'].apply(lambda x: x.year)
df['month'] = df['date'].apply(lambda x: x.month)
df['day'] = df['date'].apply(lambda x: x.day)
```

```python
df.head()
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.info()
```

# Определение наиболее значимых атрибутов

```python
corr = df.corr()


# Fill diagonal and upper half with NaNs
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True
corr[mask] = np.nan
(corr
 .style
 .background_gradient(cmap='coolwarm', axis=None, vmin=-1, vmax=1)
 .highlight_null(null_color='#f1f1f1')  # Color NaNs grey
 .set_precision(2))
```

```python
corrmatr = df.corr()
corrmatr *= np.tri(*corrmatr.values.shape, k=-1).T
corrmatr = corrmatr.stack().sort_values(ascending = False).reset_index()
corrmatr.columns = ['Признак 1', 'Признак 2', 'Корреляция']
corrmatr[(corrmatr['Корреляция'] >= 0.7) + (corrmatr['Корреляция'] <= -0.7)]
```

```python
corrmatrix = df.corr()
corrmatrix *= np.tri(*corrmatrix.values.shape, k=-1).T
corrmatrix = corrmatrix.stack().sort_values(ascending = False).reset_index()
corrmatrix.columns = ['Признак 1', 'Признак 2', 'Корреляция']
corrmatrix[(corrmatrix['Корреляция'] >= 0.7) + (corrmatrix['Корреляция'] <= -0.7)]
drop_columns = corrmatrix[(corrmatrix['Корреляция'] >= 0.82) + (corrmatrix['Корреляция'] <= -0.7)]['Признак 2']
df.drop(drop_columns, axis=1, inplace=True)
corrmatrix[(corrmatrix['Корреляция'] >= 0.7) + (corrmatrix['Корреляция'] <= -0.7)]
```
